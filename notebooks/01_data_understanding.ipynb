````python
# Check suspicious age values
df[df["age"] == 0]
````

````python
# Check unusually high late-payment values
late_payment_columns = [
    "NumberOfTime30-59DaysPastDueNotWorse",
    "NumberOfTimes90DaysLate",
    "NumberOfTime60-89DaysPastDueNotWorse"
]

for column in late_payment_columns:
    print(f"\n{column}")
    print(df[column].value_counts().sort_index().tail(10))
````

````python
suspicious_late_payments = df[
    df["NumberOfTime30-59DaysPastDueNotWorse"].isin([96, 98])
]

suspicious_late_payments[
    [
        "NumberOfTime30-59DaysPastDueNotWorse",
        "NumberOfTimes90DaysLate",
        "NumberOfTime60-89DaysPastDueNotWorse"
    ]
].value_counts()
````

````python
suspicious_mask = (
    df["NumberOfTime30-59DaysPastDueNotWorse"].isin([96, 98])
)

normal_default_rate = df.loc[
    ~suspicious_mask, "SeriousDlqin2yrs"
].mean() * 100

suspicious_default_rate = df.loc[
    suspicious_mask, "SeriousDlqin2yrs"
].mean() * 100

print(f"Normal records default rate: {normal_default_rate:.2f}%")
print(f"Suspicious records default rate: {suspicious_default_rate:.2f}%")
````

````python
df["RevolvingUtilizationOfUnsecuredLines"].quantile(
    [0.90, 0.95, 0.99, 0.999, 1.0]
)
````

````python
(df["RevolvingUtilizationOfUnsecuredLines"] > 1).sum()
````

````python
utilization_groups = pd.cut(
    df["RevolvingUtilizationOfUnsecuredLines"],
    bins=[-float("inf"), 1, 10, 100, 1000, float("inf")],
    labels=["≤1", "1-10", "10-100", "100-1000", ">1000"]
)

df.groupby(utilization_groups, observed=False)["SeriousDlqin2yrs"].agg(
    customer_count="count",
    default_rate="mean"
)
````

````python
df["DebtRatio"].quantile(
    [0.50, 0.75, 0.90, 0.95, 0.99, 0.999, 1.0]
)
````

````python
(df["DebtRatio"] > 1).sum()
````

````python
df.groupby(df["MonthlyIncome"].isna())["DebtRatio"].agg(
    customer_count="count",
    median="median",
    mean="mean",
    maximum="max"
)
````
