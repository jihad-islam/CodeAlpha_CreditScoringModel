````python
delinquency_columns = {
    "30-59 Days Late": "NumberOfTime30-59DaysPastDueNotWorse",
    "60-89 Days Late": "NumberOfTime60-89DaysPastDueNotWorse",
    "90+ Days Late": "NumberOfTimes90DaysLate"
}

for label, column in delinquency_columns.items():

    groups = pd.cut(
        df[column],
        bins=[-1, 0, 1, 2, 3, 5, 95],
        labels=["0", "1", "2", "3", "4-5", "6+"]
    )

    result = (
        df[df[column] < 96]
        .groupby(groups, observed=False)["SeriousDlqin2yrs"]
        .agg(customer_count="count", default_rate="mean")
    )

    result["default_rate"] *= 100

    print(f"\n{label}")
    display(result)
````

````python
income_groups = pd.qcut(
    df["MonthlyIncome"],
    q=5,
    duplicates="drop"
)

income_default_rate = (
    df.groupby(income_groups, observed=False)["SeriousDlqin2yrs"]
    .agg(customer_count="count", default_rate="mean")
)

income_default_rate["default_rate"] *= 100

income_default_rate
````

````python
income_missing_analysis = (
    df.groupby(df["MonthlyIncome"].isna())["SeriousDlqin2yrs"]
    .agg(customer_count="count", default_rate="mean")
)

income_missing_analysis["default_rate"] *= 100

income_missing_analysis
````

````python
income_available = df[df["MonthlyIncome"].notna()].copy()

debt_ratio_groups = pd.qcut(
    income_available["DebtRatio"],
    q=5,
    duplicates="drop"
)

debt_ratio_default_rate = (
    income_available
    .groupby(debt_ratio_groups, observed=False)["SeriousDlqin2yrs"]
    .agg(customer_count="count", default_rate="mean")
)

debt_ratio_default_rate["default_rate"] *= 100

debt_ratio_default_rate
````

````python
remaining_features = [
    "NumberOfOpenCreditLinesAndLoans",
    "NumberRealEstateLoansOrLines",
    "NumberOfDependents"
]

for column in remaining_features:
    groups = pd.qcut(
        df[column],
        q=5,
        duplicates="drop"
    )

    result = (
        df.groupby(groups, observed=False)["SeriousDlqin2yrs"]
        .agg(
            customer_count="count",
            default_rate="mean"
        )
    )

    result["default_rate"] *= 100

    print(f"\n{column}")
    display(result)
````

## EDA Summary and Key Findings

### Dataset Overview
- The dataset contains 150,000 customer records.
- The target variable is highly imbalanced:
  - 93.32% did not experience serious delinquency.
  - 6.68% experienced serious delinquency.
- Missing values are present in `MonthlyIncome` and `NumberOfDependents`.

### Data Quality Findings
- One record contains an invalid age of 0.
- The delinquency features contain anomalous values of 96 and 98.
- These anomalous delinquency records have a significantly higher default rate than normal records.
- `RevolvingUtilizationOfUnsecuredLines` contains extreme values far above the typical range.
- Extreme `DebtRatio` values are strongly associated with missing `MonthlyIncome`.
- `MonthlyIncome` is right-skewed and contains extreme high-income values.

### Default Risk Patterns
- Default risk decreases consistently with age.
- Higher revolving credit utilization is strongly associated with higher default risk.
- Previous late payments, especially 60–89 and 90+ days late, are strong indicators of default risk.
- Lower-income groups generally have higher default rates.
- Higher debt ratios are generally associated with increased default risk.
- Customers with more dependents show higher default rates.
- The relationships involving the number of credit lines and real estate loans appear non-linear.

### Modeling Considerations
- Accuracy alone will not be sufficient because of severe class imbalance.
- Precision, Recall, F1-score, and ROC-AUC should be used for model evaluation.
- Missing values require appropriate treatment before model training.
- Extreme and anomalous values should be investigated carefully rather than removed blindly.
- Feature transformations and engineered indicators may be useful for highly skewed or anomalous variables.
- Correlation alone should not be used for feature selection because several important relationships are non-linear.
